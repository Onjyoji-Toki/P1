```python
#down up
import matplotlib.pyplot as plt
import numpy as np
import os

def intg(d):
    return (d[510:525] - d[:500].mean()).sum()

def histdata(loop, farray, numbins):
    hist, edges = np.histogram(farray, bins=1000, range=(0,1000))
    return hist, edges

total_hist = np.zeros(1000)
print(total_hist)
path = '/Users/r.shinoda/p1/data'
direction = 'up'   #edit here!!!!!!
date = '01241541'   #edit here!!!!!!
data_path = os.path.join(path, direction + date, date + direction)
print(data_path)

num_of_loops = 24   #edit here!!!!!!
```

```python
for loop in range(num_of_loops):
    file_path = os.path.join(data_path +str(loop)+ '.txt')
    print(file_path)
    with open(file_path) as f:
        flines = f.readlines()
        fnum = len(flines)/1023
        print(fnum)
        ftest = np.array([int(line) for line in flines[:1023]])
        ftest.mean()
        #plt.plot(ftest - ftest[0:500].mean())
        
        fsubtracted = ftest - ftest[0:500].mean()
        fsubtracted[500:600].sum()
        #plt.xlim([500, 570])
        
        N_points = 1023
        fdata = np.array([int(line) for line in flines])
        flength = len(fdata)//N_points
        fdata_res = fdata[:flength * N_points].reshape(-1, N_points)
        farray = [-intg(d) for d in fdata_res]
        fhist, fedges = histdata(loop, farray, numbins)
        print(fedges)
        total_hist += fhist
        
        plt.figure()
        plt.plot(fedges[:-1], fhist)
        plt.xlim([0, 1000])
        plt.xlabel('energy')
        plt.ylabel('ADC counts in loop' + str(loop))
```

```python
plt.figure()
plt.plot(fedges[:-1], total_hist)
plt.xlim([0, 1000])
#plt.ylim([0, 200])
plt.xlabel('energy')
plt.ylabel('ADC counts')
plt.show()
```
