```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = r'C:\Users\kyoto\WaveDump'
date = '01241002'   #edit here!!!!!!
```

```python
edge_path = r'C:\Users\kyoto\WaveDump\edges.npy'
edges = np.load(edge_path)
#print(edges)
```

```python
num_of_loops = 27   #edit here!!!!!!
direction = 'down'   #edit here!!!!!!
direction_path = os.path.join(path, direction + date, date + direction + 'with' + str(num_of_loops) + 'loops.npy')
print(direction_path)
hist = np.load(direction_path)
```

```python
carib_path = r'C:\Users\kyoto\WaveDump\caribration'
```

```python
linapp_path = os.path.join(carib_path, direction, direction + '_linapp_params.npy')
slope, intercept = np.load(linapp_path)
print([slope, intercept])
```

```python
keV_edges_path = os.path.join(carib_path, direction, direction + '_keV_edges.npy')
keV_edges = edges*slope + intercept
np.save(keV_edges_path, keV_edges)
#print(keV_edges)
```

```python
plt.plot(keV_edges, hist)
plt.yscale('log')
plt.xlim([0, 1000*slope + intercept])
plt.xlim([700, 1500])
plt.ylim([0, 10000])
plt.xlabel('Energy [keV]')
plt.ylabel('counts')
plt.grid(True)
plt.axvline(841, color="red", linestyle="dashed", linewidth = 1)
plt.axvline(963, color="red", linestyle="dashed", linewidth = 1)
plt.axvline(1435, color="green", linestyle="dashed", linewidth = 1)
```
